```python
df = pd.DataFrame({'num_legs': [2, 2, 8, 0],
                   'num_wings': [2, 0, np.nan, 0],
                   'num_specimen_seen': [10, 2, 1, 8]},
                  index=['falcon', 'dog', 'spider', 'fish'])
```

```python
df.head()
```

```python
df.dropna() #drop NA values
```

```python
df.dropna(how='all',axis=1)
```

```python
df.drop_duplicates(subset=["num_legs"]) #removed duplicates from num_legs colums
```

```python
df.drop_duplicates(subset=["num_legs"],keep='first') #keep 1st duplicate
```

```python
df.drop_duplicates(subset=["num_legs"],keep='last')#keep last duplicate
```
