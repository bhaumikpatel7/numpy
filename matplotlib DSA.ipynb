```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
y=x*2
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,color='g',linestyle=':',linewidth=5)
plt.title('This is x vs y')
plt.xlabel('This is x ')
plt.ylabel('This is y')

plt.show()
```

```python
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.grid(True)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.subplot(1,2,2)
plt.plot(x,y2)

plt.show()
```

```python
student = {'bob':23,'julia':67,'khushu':100,'bhaumik':99}
```

```python
names=list(student.keys())
```

```python
names
```

```python
values=list(student.values())
```

```python
values
```

```python
plt.bar(names,values)
plt.show()
```

```python
plt.bar(names,values)
plt.title('Distribution of student marks')
plt.xlabel('name of student')
plt.ylabel('marks of student')
plt.show()
```

```python
plt.barh(names,values,color='g') #barh
plt.title('Distribution of student marks')
plt.xlabel('name of student')
plt.ylabel('marks of student')
plt.show()
```

```python
fruit = ['apple','mango','guava','orange','pineapple']
quantity = [12,4,3,12,10]
```

```python
plt.pie(quantity,labels=fruit)
plt.show()
```

```python
plt.pie(quantity,labels=fruit,autopct='%0.1f%%')
plt.show()
```

```python
plt.pie(quantity,labels=fruit,radius=3)

plt.show()
```

```python
plt.pie(quantity,labels=fruit,radius=2)
plt.pie([100],radius=1,colors=['white'])
plt.show()
```
